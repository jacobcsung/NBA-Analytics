The final change I want to make is to create primary keys for each of the tables. As of right now, we have unique identifiers in the ```PLAYER_ID```, ```SEASON_ID```, etc, however since each of the tables include multiple seasons as well as season types, the identifiers repeat within the same columns, and thus cannot be our primary keys. To work around this, we will be creating a new column that concatenates the identifier ID with the season ID associated with each entry. For exampl, in the ```team_df``` concatenating ```TEAM_ID``` with ```SEASON_ID``` will ensure that each row entry within our primary key column is in fact unique.


````python
temp_id = []
for i in range(len(player_df.index)):
    id1 = str(player_df['PLAYER_ID'][i])
    id2 = str(player_df['SEASON_ID'][i])
    id1 += id2
    temp_id.append(int(id1))

#Checking for Duplicates
len(temp_id) == len(set(temp_id))
````

````python
player_df['PLAYER_LABEL'] = temp_id
temp = player_df['PLAYER_LABEL']
player_df.drop(labels=['PLAYER_LABEL'], axis = 1, inplace = True)
player_df.insert(0, 'PLAYER_LABEL', temp)

player_df
````

````python
temp_id = []
for i in range(len(team_df.index)):
    id1 = str(team_df['TEAM_ID'][i])
    id2 = str(team_df['SEASON_ID'][i])
    id1 += id2
    temp_id.append(int(id1))
    
#Checking for Duplicates
len(temp_id) == len(set(temp_id))
````

````python
team_df["TEAM_LABEL"] = temp_id
temp = team_df["TEAM_LABEL"]
team_df.drop(labels=["TEAM_LABEL"], axis = 1, inplace = True)
team_df.insert(0, "TEAM_LABEL", temp)

team_df
````

\
For ```seasons_df``` we will concatenate ```SEASON_ID``` with the row index to get ```SEASON_LABEL```

````python
temp_id = []
for i in range(len(season_df.index)):
    id1 = str(i+1)
    id2 = str(season_df['SEASON_ID'][i])
    id1 += id2
    temp_id.append(int(id1))
    
#Checking for Duplicates
len(temp_id) == len(set(temp_id))
````

````python
del season_df["SEASON"]
del season_df["SEASON_TYPE"]

season_df["SEASON_LABEL"] = temp_id
temp = season_df["SEASON_LABEL"]
season_df.drop(labels=["SEASON_LABEL"], axis = 1, inplace = True)
season_df.insert(0, "SEASON_LABEL", temp)


season_df
````
